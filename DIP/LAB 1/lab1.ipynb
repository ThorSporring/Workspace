```python
from imageio import imread, imwrite # Library for reading and writing images
from pylab import * # library containing lots of basic functions for data processing
from skimage.util import img_as_ubyte, img_as_float # Convert image between interval ~(0,255) and ~(0,1)
import numpy as np # Library for handling matrices
import matplotlib.pyplot as plt # Library for creating plots
from matplotlib.pyplot import imshow # import imshow directly since we will use it a lot
plt.gray() #specify gray-scale colormap for plots

im = imread('../Images/cameraman.png')
imshow(im)
```

### Show the part of the image that consists of the rows from 50 to 100 and columns from 100 to 150.

```python
im_limited = im[50:100, 100:150];

imshow(im_limited)
```

### Use the histogram function hist() on the image cameraman.png. You will have to flatten the image (makeit a vector) to use the hist()-function. You can do this using im.flatten()

```python
im_flat = im.flatten()
hist(im_flat);
```

###  Generate a new image by setting the pixel values below the mean value in the original image to 0 and the values above the mean value are set to 255. You can use the comparison (<>) to create a binary image consisting of true/false values, e.g. im>x. Afterwards, you ca multiply by 255 and cast to uint8 (interval 0:255). Scaling and casting between oat (inteval 0:1) and uint8 can also be acomplished using using img_as_oat and img_as_ubyte from skimage.util. see https://scikit-image.org/docs/dev/user_guide/data_types.html.

```python

bin_im = (im > im.mean()) * 255

bin_im_ubyte = img_as_ubyte(bin_im)

imshow(bin_im_ubyte)
```
